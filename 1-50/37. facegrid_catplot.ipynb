### Create a facet grid (small multiples) of countplots: one small countplot per customer_segment showing the distribution of payment_method within that segment.

```python
# using Catplot

g=sns.catplot(
    data=df,
    x='payment_method',
    col='customer_segment',
    kind='count',
    col_wrap=3, # Adjusts how many plots per row
    height=6, # Height of each individual facet
    palette='mako'
)
g.set_titles("{col_name} Segment") # Renames titles from 'customer_segment = Corporate' to just 'Corporate Segment'
g.set_xticklabels(rotation=45)     # Rotates labels so they don't overlap
plt.subplots_adjust(top=0.9)       # Make room for the main title

plt.show()
```
