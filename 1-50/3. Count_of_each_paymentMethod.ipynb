```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('ecommerce_clean.csv')
```

### 3. Count how many orders were paid using each payment method (Cash on Delivery, Credit Card, etc.).

```python
df.payment_method.value_counts()
#? or

df.groupby('payment_method')['payment_method']\
    .count()\
    .sort_values(ascending=False)
```

### 3.1 Which payment method accounts for the highest volume of orders. 

```python
plt.figure(figsize=(10,5))

ax=sns.countplot(
    data=df,
    x='payment_method',
    order=df.payment_method.value_counts().index,
    hue='payment_method',
    width=0.5,
    palette='mako',
    legend=False
)  

# Add the numbers on top of the bars
for container in ax.containers:
    ax.bar_label(container,padding=1,fontsize=10)

plt.title("Distribution of Payment Methods")
plt.xlabel("Payment Methods")
plt.ylabel('Frequency')

plt.show()
```
