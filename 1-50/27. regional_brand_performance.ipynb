```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
pd.set_option('display.max_columns',None)
df=pd.read_csv(r'D:\Data Analytics\Python\Panda\Problem_Solving\ecommerce_clean.csv')
```

```python
df.head(2)
```

### Calculate the average profit_margin for each brand-region combination, Then create a heatmap of this pivot table.

```python
region_brand=pd.crosstab(
    index=df.brand,
    columns=df.region,
    values=df.profit_margin,
    aggfunc='mean',
    )

region_brand=round(region_brand,3)
```

### Visualization

```python
plt.figure(figsize=(8,10))

sns.heatmap(
    data=region_brand,
    annot=True,
    fmt='1.1f',
    cmap="RdYlGn",
    linewidths=0.6,
    cbar_kws={'label': 'Avg Profit Margin (%)'}
)


plt.title('Regional Brand Performance: Profit Margin Analysis', fontsize=15, pad=20)
plt.xlabel('Region')
plt.ylabel('Brand')
plt.show()
```
