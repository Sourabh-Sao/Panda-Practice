```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',None)
```

```python
df=pd.read_csv(r'D:\Data Analytics\Python\Panda\Problem_Solving\ecommerce_clean.csv')
```

### 7. Count how many orders were placed on each day of the week

```python
df.groupby('order_day_of_week')['order_id'].agg(order_count='count').sort_values(by='order_count', ascending=False)
```

### Visualization


```python
# The primary purpose of countplot is to visualize raw frequencies (counts) of each category.

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.figure(figsize=(10,8))

ax=sns.countplot(
    data=df,
    x='order_day_of_week',
    hue='order_day_of_week',
    palette='vlag',
    width=0.5,
    order=day_order
)

for container in ax.containers:
    ax.bar_label(container,padding=1)

plt.title('Weekly Order Distribution')
plt.xlabel("Day of Week")
plt.ylabel("Order Count")

plt.show()
```
