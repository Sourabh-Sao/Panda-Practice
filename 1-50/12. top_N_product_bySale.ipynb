```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',None)
df=pd.read_csv(r'D:\Data Analytics\Python\Panda\Problem_Solving\ecommerce_clean.csv')
```

```python
df.head(5)
```

### Find the Top 10 product_names with the highest total sales_amount. Show product name and total sales in a sorted table.

```python
df.groupby('product_name')['sales_amount']\
    .agg(total_sale='sum')\
    .nlargest(n=10,columns='total_sale',keep='first')\
    .reset_index()

# or 

df.groupby('product_name')['sales_amount']\
    .agg(total_sale='sum')\
    .sort_values(by='total_sale',ascending=False).head(10)\
    .reset_index()
```

### Visualization

```python
top_n=df.groupby('product_name')['sales_amount']\
    .agg(total_sale='sum')\
    .nlargest(n=10,columns='total_sale',keep='first')\
    .reset_index()

plt.figure(figsize=(6,4))

sns.barplot(
    data=top_n,
    x='total_sale',
    y='product_name',
    palette='Blues'
)

plt.grid(alpha=0.3)

plt.title('Top 10 Products by Sale')
plt.show()
```
