### Top 3 products (by total sales_amount) in each category. Show category, product_name, total sales, and rank within category.

```python
prod=df.groupby(['category','product_name'])\
    .agg(total_sale=('sales_amount','sum'))\
    .sort_values(by=['category','total_sale'],ascending=[True,False])

prod['rnk']=prod.groupby('category')['total_sale']\
    .rank(method='min',ascending=False).astype('int')

prod[prod['rnk']<=3].reset_index()
```
