```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
pd.set_option('display.max_columns',None)
df=pd.read_csv(r'D:\Data Analytics\Python\Panda\Problem_Solving\ecommerce_clean.csv')
```

### Investigate if higher discount_percentage leads to higher quantity ordered. Create a scatter plot of discount_percentage vs quantity, add a trend line (regression line), and color points by category.

```python
sns.lmplot(
    data=df,
    x='discount_percentage',
    y='quantity',
    hue='category',
    aspect=1.5,
    height=6,
    line_kws={'linewidth':3},
    scatter_kws={'alpha':0.4, 's':50}
)

plt.title('Impact of Discount % on Order Quantity', fontsize=16, pad=20)
plt.xlabel('Discount Percentage (%)', fontsize=12)
plt.ylabel('Quantity Ordered', fontsize=12)

plt.show()
```
