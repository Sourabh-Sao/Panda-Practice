### 2. How many unique customers are in this dataset? How many unique products?

```python
# len() - This includes NaN as a unique value in the count.
print("Unique Customer Count : ",len(df.customer_id.unique()))
print("Unique Product Count : ",len(df.product_id.unique()))

#? or
# This excludes NaN values by default
df.customer_id.nunique()
df.product_id.nunique()
```
