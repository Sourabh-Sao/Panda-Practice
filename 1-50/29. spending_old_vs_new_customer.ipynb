```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
pd.set_option('display.max_columns',None)
df=pd.read_csv(r'D:\Data Analytics\Python\Panda\Problem_Solving\ecommerce_clean.csv')
```

### Compare total_amount distribution between repeat and non-repeat customers using side-by-side boxplots or violin plots. Add a title and make it publication-ready (labels, size).

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=df,
    x='is_repeat_customer',
    y='total_amount',
    hue='is_repeat_customer',
    palette='Set2',
    showmeans=True,
)

plt.title('Spending Distribution: Repeat vs. Non-Repeat Customers', fontsize=16, pad=20)
plt.xlabel('Repeat Customer Status', fontsize=12)
plt.ylabel('Order Total Amount ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.violinplot(
    data=df,
    x='is_repeat_customer',
    y='total_amount',
    hue='is_repeat_customer',
    palette='Set2',
)

plt.title('Spending Distribution: Repeat vs. Non-Repeat Customers', fontsize=16, pad=20)
plt.xlabel('Repeat Customer Status', fontsize=12)
plt.ylabel('Order Total Amount ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```
