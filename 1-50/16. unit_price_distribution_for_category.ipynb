```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',None)
df=pd.read_csv(r'D:\Data Analytics\Python\Panda\Problem_Solving\ecommerce_clean.csv')
```

### Show a boxplot of unit_price distribution for the top 6 categories (by number of orders).

```python
top6_category=df['category'].value_counts().nlargest(6).index
top6=df[df.category.isin(top6_category)]
```

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    data=top6,
    x='category',
    y='unit_price',
    order=top6_category,
    palette='Set2',
    hue='category'
)

plt.title('Unit Price Range for Top 6 Most Popular Categories')
plt.xlabel('Category')
plt.ylabel('Unit Price ($)')

plt.show()
```
