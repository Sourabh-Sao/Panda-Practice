```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',None)
df=pd.read_csv(r'D:\Data Analytics\Python\Panda\Problem_Solving\ecommerce_clean.csv')
```

### Show monthly total sales_amount trend (order_month 1â€“12) separately for each region (use color/line for each region).

```python
g=df.groupby(['region','order_month'])['sales_amount'].agg(total_sale='sum')
g=g.unstack(level=0)
```

### Visualization
Barplot

```python
plt.figure(figsize=(12,6))


sns.barplot(
    data=df,
    x='order_month',
    y='sales_amount',
    estimator='sum',
    hue='region',
    errorbar=None,
    palette='magma',
    # marker='o',
)

plt.title('Monthly Sales Trend by Region')
plt.xticks(range(0,13))
plt.grid(alpha=0.5,linestyle='--')

plt.legend(
    bbox_to_anchor=(1,1.05),
    title='Regions'
)

plt.show()
```

Lineplot

```python
plt.figure(figsize=(12,6))


sns.lineplot(
    data=df,
    x='order_month',
    y='sales_amount',
    estimator='sum',
    hue='region',
    errorbar=None,
    palette='Set2',
    marker='o',
)

plt.title('Monthly Sales Trend by Region')
plt.xticks(range(0,13))
plt.grid(alpha=0.5,linestyle='--')

plt.legend(
    bbox_to_anchor=(1,1.05),
    title='Regions'
)

plt.show()
```
