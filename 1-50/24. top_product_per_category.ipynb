### For each category, find the product with the highest average rating among products. Show category, best product name, avg rating, and number of orders for that product.

```python
# Approach 1 : Using groupby + sort_values + head(1)
prod_stat=df.groupby(['category','product_name']).agg(avg_rating=('rating','mean'),order_count=('order_id','count'))\
    .reset_index()

prod_stat=prod_stat.sort_values(by=['category','avg_rating','order_count'], ascending=[True,False,False])

prod_stat=prod_stat.groupby('category').head(1)
```

```python
# Approach 2: Using groupby + idxmax()

p_stat=df.groupby(['category','product_name']).agg(avg_rating=('rating','mean'),order_count=('order_id','count'))\
    .reset_index()

idx=p_stat.groupby('category')['avg_rating'].idxmax()
p_stat.loc[idx]
```

```python
h_cat=df.groupby('category')['sales_amount'].sum().reset_index()

max_idx=h_cat['sales_amount'].idxmax()
h_cat.loc[max_idx]
```
