```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',None)
df=pd.read_csv(r'D:\Data Analytics\Python\Panda\Problem_Solving\ecommerce_clean.csv')
```

### 11.Create a new column called net_amount = total_amount - tax_amount - shipping_cost. Then show the average net_amount for each combination of customer_gender and customer_segment.

```python
df['net_amount']=df['total_amount']-(df['tax_amount']+df['shipping_cost'])
df
```

```python
df.groupby(['customer_gender','customer_segment'])['net_amount']\
    .agg(avg_net_amount='mean').reset_index()
```

### Visualization

```python
plt.figure(figsize=(8,4))

ax=sns.barplot(
    data=df,
    x='customer_segment',
    y='net_amount',
    hue='customer_gender',
    errorbar=None,
    palette='mako',
    width=0.5
)
plt.legend(
    bbox_to_anchor=(1,1.05),
    title='Gender'
    )
plt.title('Average Net Revenue by Segment and Gender')

plt.show()
```

```python
df.drop(columns='net_amount',inplace=True)
```

### Who is eating our margins?" Calculate the "Deduction Ratio"what percentage of the total price is being lost to taxes and shipping?

### Which customer segment has the highest overhead costs (shipping + tax), and does this significantly reduce their net value compared to other segments?

```python
df.head(2)
```

```python
df['overhead']=df.tax_amount+df.shipping_cost
df['overhead_ratio']=round((df.overhead/df.total_amount)*100,2)

plt.figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x='customer_segment',
    y='overhead_ratio',
    hue='customer_gender',
    palette='Set2'
)
plt.title("Overhead Leakage: % of Total Amount lost to Tax & Shipping")
plt.ylabel('overhead ratio (%)')
plt.show()
```

```python
df.drop(columns=['overhead','overhead_ratio'],inplace=True)
```
