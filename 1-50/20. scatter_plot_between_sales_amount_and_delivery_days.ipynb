```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',None)
df=pd.read_csv(r'D:\Data Analytics\Python\Panda\Problem_Solving\ecommerce_clean.csv')
```

### Make a scatter plot of delivery_time_days vs sales_amount, colored by order_status ("Delivered", "Cancelled", "Returned", etc.). Add a title and make figure bigger.

```python
plt.figure(figsize=(10,8))

sns.scatterplot(
    data=df,
    x='delivery_time_days',
    y='sales_amount',
    hue='order_status',
    alpha=0.7,
    s=100
)

plt.legend(bbox_to_anchor=(1.20,1))

plt.title('Sales Amount vs. Delivery Time (Colored by Order Status)', fontsize=15)
plt.xlabel('Delivery Time (Days)', fontsize=12)
plt.ylabel('Sales Amount ($)', fontsize=12)
plt.show()
```
