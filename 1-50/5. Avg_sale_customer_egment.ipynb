```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',None)
```

```python
df=pd.read_csv(r'D:\Data Analytics\Python\Panda\Problem_Solving\ecommerce_clean.csv')
```

### 5. Calculate the average sales_amount for each customer_segment ("New", "Regular", "Loyal", etc.).

```python
df.groupby('customer_segment')['sales_amount'].agg(avg_sales='mean').reset_index()
```

### Visualization

```python
plt.figure(figsize=(10,6))
segment_avg=df.groupby('customer_segment')['sales_amount'].agg(Avg_Sales='mean').reset_index()

sns.barplot(
    data=segment_avg,
    x=segment_avg.customer_segment,
    y=segment_avg.Avg_Sales,
    color='lightblue',
    width=0.5
)

plt.title('Average Sales per Customer Segment')
plt.show()
```

### Alternate way

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=df,
    x=df.customer_segment,
    y=df.sales_amount,
    estimator='mean',
    palette='Blues',
    hue=df.customer_segment,
    errorbar=None,
    width=0.5
)

plt.title('Average Sales per Customer Segment')
plt.show()
```
