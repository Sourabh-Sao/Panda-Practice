### 4. Find out how many orders were placed in each order_month. Sort the result from highest to lowest number of orders.

```python
df.groupby('order_month')['order_id'].count() \
    .sort_values(ascending=False) \
    .reset_index()

#? or

df.groupby('order_month')['order_id'].agg(order_count='count')\
    .sort_values(by='order_count',ascending=False)\
    .reset_index()
```
