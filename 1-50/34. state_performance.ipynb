### Task: For each state, calculate:

- total orders
- average delivery_time_days
- % of orders with discount_percentage > 0

Then sort states by total orders descending and show only the top 10 in a styled table (highlight the state with highest avg delivery time in red).

```python
state_data=df.groupby('state')\
    .agg(
        order_count=('order_id','count'),
        avg_delivery_time_days=('delivery_time_days','mean'),
        discount_pct=('discount_percentage',lambda x:(x>0).mean()*100)
        )
    
state_data.nlargest(n=10,columns='order_count')
```
