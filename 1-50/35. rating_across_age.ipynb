### Show how average rating changes across customer_age (binned in 10-year groups) separately for male and female customers. Use a line plot with hue=gender and confidence interval bands.

```python
df['age_bucket'] = pd.cut(df['customer_age'], bins=range(10, 91, 10), right=False)

# 2. Convert categories to strings (prevents Seaborn plot errors with Interval objects)
df['age_bucket_str'] = df['age_bucket'].astype(str)

# 3. Plot using Seaborn
# We pass the full dataframe (df) so Seaborn can calculate the Confidence Interval
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=df, 
    x='age_bucket_str', 
    y='rating', 
    hue='customer_gender', 
    marker='o', 
)

plt.title('Average Rating by Age Group and Gender (with 95% CI)')
plt.xlabel('Age Group')
plt.ylabel('Average Rating')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
